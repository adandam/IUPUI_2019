```python
##Outcome: Premature Deaths

#SQL connection imports
import MySQLdb

#plotly imports
import plotly.plotly as py
import plotly.figure_factory as ff

#general data imports
import numpy as np
import pandas as pd

#connection
conn = MySQLdb.connect(host="localhost", user="i501fa18_group5", passwd="*85E!4LfLQbQ", db="i501fa18_group5")
cursor = conn.cursor()
sql = "SELECT avg(distinct(premature_deaths)/population_estimate_value)*100, state FROM CombinedData group by state"
df = pd.read_sql(sql, con=conn)

for col in df.columns:
    df[col] = df[col].astype(str)

scl = [[0.0, 'rgb(247,240,240)'],[0.2, 'rgb(235,218,218)'],[0.4, 'rgb(220,188,189)'],
            [0.6, 'rgb(208,154,154)'],[0.8, 'rgb(177,117,107)'],[1.0, 'rgb(143,84,39)']]

df['text'] = 'State: ' + df['state'] + '<br>' + 'Average % Premature Deaths: '+df['avg(distinct(premature_deaths)/population_estimate_value)*100']

data = [ dict(
        type='choropleth',
        colorscale = scl,
        autocolorscale = False,
        locations = df['state'],
        z = df['avg(distinct(premature_deaths)/population_estimate_value)*100'].astype(float),
        locationmode = 'USA-states',
        text = df['text'],
        marker = dict(
            line = dict (
                color = 'rgb(255,255,255)',
                width = 2
            ) ),
        colorbar = dict(
            title = "Average Percent of Premature Deaths")
        ) ]

layout = dict(
        title = 'Average Premature Deaths in Each State',
        geo = dict(
            scope='usa',
            projection=dict( type='albers usa' ),
            showlakes = True,
            lakecolor = 'rgb(255, 255, 255)'),
             )
    
fig = dict( data=data, layout=layout )
py.iplot( fig, filename='d3-cloropleth-map' )
```
